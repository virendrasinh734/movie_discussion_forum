```python
import numpy as np 
import pandas as pd 
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics.pairwise import cosine_similarity

```

```python
data_movie = pd.read_csv("archive/movie.csv")
data_rating = pd.read_csv("archive/rating.csv")
```

```python
data_movie.head()
```

```python
data_rating.head()
```

```python
movie = data_movie.loc[:,["movieId","title"]]
rating = data_rating.loc[:,["userId","movieId","rating"]]
```

```python
data = pd.merge(movie,rating)
data = data.iloc[:1000000,:]
```

```python
user_movie_table = data.pivot_table(index = ["title"],columns = ["userId"],values = "rating").fillna(0)
```

```python
user_movie_table
```

```python
cosine_sim_matrix = cosine_similarity(user_movie_table)
```

```python
cosine_sim_matrix
```

```python
import pickle
with open('user_movie_table.pkl', 'wb') as f:
    pickle.dump(user_movie_table, f)

# Store cosine similarity matrix in a pickle file
with open('cosine_similarity_matrix.pkl', 'wb') as f:
    pickle.dump(cosine_sim_matrix, f)
```
